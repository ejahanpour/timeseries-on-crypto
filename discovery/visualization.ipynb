```python
import os, datetime
import pandas as pd
import matplotlib.pyplot as plt
import mplfinance as fplt
```

```python
token_dir = '/mnt/intermediate_datasets/'
token_list = os.listdir(token_dir)
```

```python
sample_token = 'BTC.csv'
token_df = pd.read_csv(f'{token_dir}/{sample_token}', index_col = 'timestamp').drop('Unnamed: 0', axis = 1)
token_df.index = token_df.index.map(datetime.datetime.fromtimestamp)
```

```python
token_df.head()
```

```python
# token_df.iloc[:, 2].plot()
plt.plot(token_df.loc[:, 'Open'], label = 'Open', color = 'blue')
plt.plot(token_df.loc[:, 'Close'], label = 'Close', color = 'black')
plt.plot(token_df.loc[:, 'High'], label = 'High', color = 'red')
plt.plot(token_df.loc[:, 'Low'], label = 'Low', color = 'green')
plt.legend()
plt.show()
```

```python
# token_df.iloc[:, 2].plot()

plt.plot(token_df.iloc[-30:, 2], label = 'Open', color = 'blue')
plt.plot(token_df.iloc[-30:, 3], label = 'High', color = 'black')
plt.plot(token_df.iloc[-30:, 4], label = 'Low', color = 'red')
plt.plot(token_df.iloc[-30:, 5], label = 'Close', color = 'green')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (15, 10))
fplt.plot(
            token_df[-100:],
            type='candle',
            style = 'binance',
            title='candle plot',
            ylabel='Price ($)',
            volume = True,
            show_nontrading=True,
            figratio=(10,5),
#             mav = (2, 4, 6)
        )
```

```python
plt.figure(figsize = (15, 10))
fplt.plot(
            token_df[-100:],
            type='candle',
            style = 'binance',
            title='candle plot with 4 and 6 moving average',
            ylabel='Price ($)',
            volume = True,
            show_nontrading=True,
            figratio=(10,5),
            mav = (4, 6)
        )
```
